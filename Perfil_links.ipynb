```python
import time
import requests
import json
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from requests.exceptions import ConnectionError
import time
import pymongo
from pymongo import MongoClient

client = MongoClient("52.25.140.201",27017)
db = client['ProjectCorruption']
print db
collection = db['Links_to_articles']
print collection
```

```python
total_added = 0

#list_of_keywords = ['corrupcion','corrupto','corrupta','cohecho','soborno','coima','"enriquecimiento ilicito"',\
#                   '"conflicto de interes"','nepotismo','fraude','"compra de votos"','clientelismo']
list_of_keywords = ['nepotismo','fraude','"compra de votos"','clientelismo']

chrome = webdriver.Chrome()

perfil_url = "http://www.perfil.com"

for keyword in list_of_keywords:

    pages = 0
    chrome.get('http://www.perfil.com')
    chrome.implicitly_wait(10)
    
    page = chrome.page_source
    
    search_form = chrome.find_element_by_name("search")
    search_form.send_keys(keyword)
    search_click = chrome.find_element_by_id("buscar-boton")
    search_click.send_keys(Keys.RETURN)

    while True:
        print "Keyword:"+keyword

        page = chrome.page_source
        soup = BeautifulSoup(page)

        a = soup.find_all(class_='nota busqueda-resultado')

        article_list = []
        for each in soup.find_all(class_='nota busqueda-resultado'):
            try:
                link = each.header.h1.a['href']
            except AttributeError:
                pass
            except TypeError:
                pass
            else:
                article_list.append(link)

        for each_art in article_list:
            print "Page:"+each_art

            #response = requests.get(each_search)
            if each_art[0] == '/':
                article_link = 'http://www.perfil.com'+each_art
            else:
                article_link = each_art

            exist = collection.find_one({"link": article_link, "source":"Perfil"})
            if exist == None:
                article_source = "Perfil"
                article_object = {"link":article_link, "source":article_source}
                article_id = collection.insert_one(article_object)
                total_added += 1
                print 'Link added:',article_link
                print 'Total added:',str(total_added)
            else:
                print 'Link passed:',article_link

        search_next_page = chrome.find_element_by_id("mas-resultados")
        continue_link = chrome.find_element_by_link_text('Siguiente')
        continue_link.click()
        pages += 1
        if pages >= 49:
            break
    
```
