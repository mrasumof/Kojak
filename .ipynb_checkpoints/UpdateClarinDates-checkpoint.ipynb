```python
from datetime import datetime

def convert_date(a):
    b = a.split('\n')
    for each in b:
        c = each.strip()
        if 'CalcularTimestamp(' in each:
            if c[0] != '*':
                if c[0:3] == 'var':
                    d = each.split(",")
                    time1 = datetime.fromtimestamp(float(d[1]))
                    return time1
                else:
                    return False
            else:
                return False
        else:
            return False

changed = 0

total_recs = coll_parsed.find({"source":"Clarin","parsed":"N"})

for each in coll_parsed.find({"source":"Clarin", "parsed":"N"}):
    article_link = each['link']
    response = requests.get(article_link) 
    page = response.text
    soup = BeautifulSoup(page)
    a = soup.find_all(class_="breadcrumb")[0].find(type='text/javascript')
    article_date = convert_date(a)
    if article_date != False:
        ok_id = coll_parsed.update_one({"link":article_link},{"$set" : {"date":article_date, \
                                                                        "parsed":"Y"}})
    changed +=1
    print "Processed: "+str(changed)+" of "+str(total_recs)
```
